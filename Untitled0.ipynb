```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ========================
# Create portfolio dataset
# ========================

data = {
    "Date": ["2024-01-01","2024-01-02","2024-01-03","2024-02-01","2024-02-05",
             "2024-03-01","2024-03-10","2024-03-15","2024-04-01","2024-04-10"],
    "Product": ["Phone","Laptop","Phone","Tablet","Laptop",
                "Phone","Tablet","Laptop","Phone","Laptop"],
    "Sales": [200,500,300,150,700,400,200,800,350,900]
}

df = pd.DataFrame(data)

# ========================
# Data Cleaning
# ========================

df.drop_duplicates(inplace=True)
df["Date"] = pd.to_datetime(df["Date"])

# ========================
# Analysis
# ========================

total_sales = df["Sales"].sum()
best_product = df.groupby("Product")["Sales"].sum()

print("Total Sales:", total_sales)
print("\nSales by Product:\n", best_product)

# ========================
# Monthly Trend
# ========================

df["Month"] = df["Date"].dt.month
monthly_sales = df.groupby("Month")["Sales"].sum()

plt.figure(figsize=(8,5))
monthly_sales.plot(kind="line", marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()

# ========================
# Visualization
# ========================

plt.figure(figsize=(8,5))
sns.barplot(x=best_product.index, y=best_product.values)
plt.title("Sales by Product")
plt.show()

# ========================
# Insights
# ========================

print("\nKey Insights:")
print("Best selling product:", best_product.idxmax())
print("Highest sales month:", monthly_sales.idxmax())
```
